```python
data.dropna(subset=['Sale Price','Market Price'],inplace=True)
data.info()
data
```

**Cleaning Sale Price & Market Price**


```python
def clean_price(column_name):
  data[column_name] = data[column_name].str.strip().str.split('₹').str[-1]
  data[column_name] = data[column_name].str.replace(',', '')
  data[column_name] = pd.to_numeric(data[column_name])

clean_price('Sale Price')
clean_price('Market Price')

data
```

**Cleaning Rating & Review**

```python
def clean_rating_review(column_name):
    # Convert to string, strip spaces, split by space, and take the first part (number)
    data[column_name] = data[column_name].str.strip().str.split(' ').str[0]

    # Remove commas
    data[column_name] = data[column_name].str.replace(',', '')

    # Convert to numeric, keeping NaNs for invalid values
    data[column_name] = pd.to_numeric(data[column_name])

    # Fill NaN values with 0
    data[column_name] = data[column_name].fillna(0).astype(int)

# Clean 'No_of_rating' and 'No_of_review'
clean_rating_review('No_of_rating')
clean_rating_review('No_of_review')

data
```

**Cleaning Discount Offered**

```python
def clean_discount(column_name):
    # Convert to string, strip spaces, split by '%', and take the first part (number)
    data[column_name] = data[column_name].str.strip().str.split('%').str[0]

    # Convert to numeric, keeping NaNs for invalid values
    data[column_name] = pd.to_numeric(data[column_name])

    # Fill NaN values with 0 and convert to int
    data[column_name] = data[column_name].fillna(0).astype(int)

# Clean 'Discount Offered' column
clean_discount('Discount Offered')

data

```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
data.to_csv('/content/FinalCleanedData.csv')
```
